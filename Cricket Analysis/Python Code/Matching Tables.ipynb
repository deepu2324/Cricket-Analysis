```python
import requests
import bs4
import lxml
import pandas as pd
```

```python
bat = pd.read_csv('Project CSV Files/Batting Data World Cup.csv')
m = pd.read_csv('Project CSV Files/Score Card World Cup.csv')
bowl = pd.read_csv('Project CSV Files/Bowling Data World Cup.csv')
player = pd.read_csv('Project CSV Files/dim_players.csv')
```

```python
df1 = pd.DataFrame(bat)
df2 = pd.DataFrame(m)
df3 = pd.DataFrame(bowl)
df4 = pd.DataFrame(player)
```

```python
match_ids_dict = {}
for index, row in df2.iterrows():
    key = row['Team1'] + ' vs ' + row['Team2']
    match_ids_dict[key] = row['Score Card']
```

```python
df1['Match ID'] = df1['Match'].map(match_ids_dict)
```
