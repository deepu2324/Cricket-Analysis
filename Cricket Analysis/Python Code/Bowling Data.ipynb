```python
import requests
import bs4
import lxml
import pandas as pd
```

```python
res = requests.get('https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338')
soup = bs4.BeautifulSoup(res.text , 'lxml')
frame = soup.find ('section' , id = 'main-container')
tbody = frame.find('tbody')
```

```python
m=0
t20ball = []
while m<(len(tbody))-1:
    tr = tbody.select('tr')[m]
    t1 = tr.select('span')[0]
    t2 = tr.select('span')[1]
    win = tr.select('span')[2]
    link = tr.select('a')[1]['href']
    res1 = requests.get(f'https://www.espncricinfo.com{link}')
    soup2 = bs4.BeautifulSoup(res1.text , 'lxml')
    frame2 = soup2.find('div' , id='__next')
    section2 = frame2.find('section' , id='main-container')
    #batting = []
    k=1
    while k<80:
        if (((section2.select('tr')[k]).text) == ''):
            pass
        else:
            tr1 = section2.select('tr')[k]
            j=0
            temp = []
            vs = t1.text + ' vs ' + t2.text
            temp.append(vs)
            temp.append(win.text)
            while j < (len(tr1.select('td'))):
                    td1 = tr1.select('td')[j]
                    temp.append(td1.text)
                    j+=1
            t20ball.append(temp)
        k+=1
    #t20bat.append(batting)
    m+=1
t20ball
```

```python
df = pd.DataFrame(t20ball)
df.columns = ['Match' , 'Winning Team' , 'Player' , 'Dismissal' ,  'Runs1' ,'Balls' , 'M' ,'4s' , '6s' , 'SR' , 'k' , 'l' , 'm'  ]
df.dropna( inplace = True)
```

```python
df.rename(columns = {'Dismissal' : 'Overs' ,'Balls' : 'Runs' ,  'M' :'Wickets' , '4s' : 'Econ' , 'l' : 'Wide' , 'm':'NB'} , inplace = True)
```
