```python
import requests
import bs4
import lxml
import pandas as pd
```

```python
res = requests.get('https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338')
soup = bs4.BeautifulSoup(res.text , 'lxml')
frame = soup.find ('section' , id = 'main-container')
tbody = frame.find('tbody')
```

```python
m=0
urls = []
name = []
while m<(len(tbody))-1:
    tr = tbody.select('tr')[m]
    link = tr.select('a')[1]['href']
    res1 = requests.get(f'https://www.espncricinfo.com{link}')
    soup2 = bs4.BeautifulSoup(res1.text , 'lxml')
    frame2 = soup2.find('div' , id='__next')
    section2 = frame2.find('section' , id='main-container')
    for h in section2.findAll('td'):
        a = h.find('a')
        try: 
            if 'href' in a.attrs:
                url = a.get('href')
                urls.append(url)
        except: 
            pass
        try:
            if 'title' in a.attrs:
                temp = a.get('title')
                name.append(temp)
        except:
            pass
    m+=1
urls
```

```python
urls1 = []
for i in urls:
    if i not in urls1:
        urls1.append(i)
```

```python
pname = []
for i in name:
    if i not in pname:
        pname.append(i)
```

```python
df1 =pd.DataFrame(pname)
```

```python
info = []
for i in urls1:
    try:
        res2 = requests.get(f'https://www.espncricinfo.com{i}')
        soup3 = bs4.BeautifulSoup(res2.text , 'lxml')
        temp =[]
        try:
            name = soup3.select('p')[41]
            age = soup3.select('p')[45]
            Batstyle = soup3.select('p')[47]
            Bowlstyle = soup3.select('p')[49]
            Playingrole = soup3.select('p')[51]
            temp.append(name.text)
            temp.append(age.text)
            temp.append(Batstyle.text)
            temp.append(Bowlstyle.text)
            temp.append(Playingrole.text)
            info.append(temp)
        except:
            pass
    except:
        pass
info
```

```python
df = pd.DataFrame(info)
```

```python
df.columns = ['Name' , 'Age' , 'Batting Style' , 'Bowling Style' ,  'Playing Role']
```

```python
df.to_csv("Players.csv" , index = False)
```
